```python
#Heat of a capacity of a solid
import math
from matplotlib import pyplot as plt



V=0.001
ro=6.022*(10**(28))
boltzman=1.38*(10**(-23))
theta_d=428
def f(x):
    return (((x**4)*(math.exp(x)))/(((math.exp(x)-1)**2)))

for T in range(5,501):
    def cv(T):    
        N=1000
        a=0
        b=428/T
        h=(b - 0)/N
        s= (0.5)*f(b)  
        for k in range (1,N):
            s+=f(a+k*h)
            Cv= 9*V*ro*boltzman*((T/theta_d)**3)*(s*h)
        return Cv
    CV= cv(T)
    plt.scatter(T,CV)
    plt.xlabel("Temperature (T)")
    plt.ylabel("Heat Capacity (Cv)")
```
